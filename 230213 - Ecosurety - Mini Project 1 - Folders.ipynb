```python
import pandas as pd
import os
import glob
```

```python
file_path = r'C:\Users\Levap\Desktop\Python\Ecosurety - Mini Project 1'
files = glob.glob(os.path.join(file_path, "*.xlsx")) #loading all files with ".xlsx"

df = pd.concat((pd.read_excel(x) for x in files), ignore_index=True) #concat all files
df = df[['Part Number','Product Name','QTY']]
df
```

```python
df = df.groupby(by=['Part Number','Product Name'])['QTY'].sum().reset_index()
df
#Grouping by
```
