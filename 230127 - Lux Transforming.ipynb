```python
import pandas as pd
df = pd.read_excel(r'Luxembourg - Original.xlsx')             
```

```python
#renaming Produt description
df.rename(columns={"Produt description":"Product description"},inplace=True)
```

```python
df.groupby(by = 'Product Code').count()
df.head(5)
```

```python
df['Occurence'] = df.groupby(by='Product Code')['Product Code'].transform('count')
```

```python
df['Numbering'] = df.groupby('Product Code').cumcount().add(1)
df.head(2)
#assigning a number for a each duplicate row starting with '1'
```
